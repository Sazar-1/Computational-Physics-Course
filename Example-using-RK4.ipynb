```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

# Define the stiff ODE
def stiff_ode(x, y, k=2.2067e-12, T_emb=300):
    return -k * (y**4 - T_emb**4)

# Initial condition
y0 = 1200

# Time span (adjust based on your problem)
x_span = (0, 480)

# Solve using methods suitable for stiff ODEs
methods = ['RK45','Radau', 'BDF', 'LSODA']  # Recommended for stiff problems

plt.figure(figsize=(12, 8))

for method in methods:
    sol = solve_ivp(stiff_ode, x_span, [y0], method=method, 
                   dense_output=True, rtol=1e-6, atol=1e-8)
    
    x_plot = np.linspace(*x_span, 1000)
    y_plot = sol.sol(x_plot)
    
    plt.plot(x_plot, y_plot.T, label=f'{method} method')

plt.xlabel('x')
plt.ylabel('y(x)')
plt.title('Comparison of Stiff ODE Solvers\n$dy/dx = -k(y^4 - T_{emb}^4)$')
plt.legend()
plt.grid(True)
plt.show()
```
