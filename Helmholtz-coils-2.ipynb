## The magnetic field for the __Helmholtz coils__ when the separation between the two coils __S__ is changed from 0.25R to 5R where R the radius of the coils


```python
#Import the modules that we need:
# Holmholtz coil
import numpy as np
import pylab as plt
#Define some constants:

mu0 = 4*np.pi*1e-7 #4 pi

I = 0.5

R = 0.3
#Define the values on the z axis, from -2R to +2r, in 100 increments

z = np.linspace(-7*R,7*R,100)
x=.25
while x<=7:
#Determine the magnetic field from the coils at those values of z
#The coil at z=-R/2:

    B1 = (mu0*I)/2 * R**2/((R**2+(z+x*R)**2)**(3/2))

#The coil at z=+R/2:

    B2 = (mu0*I)/2 * R**2/((R**2+(z-x*R)**2)**(3/2))

#The sum:

    B = B1 + B2

#Make the plot


#plt_1=p1t.figure(figsize=(10,6))
   # print('figure where z=',x,'R')
    
    plt_1 = plt.figure(figsize=(6, 3))
    plt_1.suptitle('figure where z=+/- '+ str(x)+'R')
    plt.plot(z,B1,label='Coil at z=-'+str(x)+'R')

    plt.plot(z,B2,label='Coil at z=+'+str(x)+'R')

    plt.plot(z,B,label='Total')

    plt.legend()

    plt.xlabel('z position [m]')

    plt.ylabel('Magnetic field [T]')

    plt.show()
    x=x+0.25
```
