## **Answer Key**  

### **Section A**  
1. **Central Difference:**  
   a) $ f'(x) \approx \frac{f(x+h) - f(x-h)}{2h} $  
   b) $ f'(0.5) \approx \frac{\sin(0.6) - \sin(0.4)}{0.2} \approx 0.87758 $  
   c) Exact: $ \cos(0.5) \approx 0.87758 $, error = 0.  

2. **Trapezoidal Rule:**  
   a) $ \int_a^b f(x)dx \approx \frac{h}{2} \left[ f(x_0) + 2\sum_{i=1}^{n-1} f(x_i) + f(x_n) \right] $  
   b) With $ h = 0.25 $:  
   $$
   \approx \frac{0.25}{2} \left[ e^0 + 2(e^{-0.25^2} + e^{-0.5^2} + e^{-0.75^2}) + e^{-1^2} \right] \approx 0.7469
   $$  
   c) Richardson extrapolation combines results from different $ h $ to cancel lower-order errors.  

### **Section B**  
3. **Gaussian Elimination:**  
   - Forward elimination yields:  
   $$
   \begin{cases}
   2x + 3y - z = 5 \\
   0x - 7y + 4z = -13 \\
   0x + 0y + \frac{50}{7}z = \frac{100}{7}
   \end{cases}
   $$  
   - Solution: $ x = 1, y = -1, z = 2 $.  

4. **Matrix Properties:**  
   a) $ \det(A) = 50 $.  
   b) A high condition number indicates sensitivity to input errors.  

### **Section C**  
5. **Euler's Method:**  
   a) $ y_{n+1} = y_n + h(-2y_n) $.  
   b) $ y(0.1) = 0.8 $, $ y(0.2) = 0.64 $.  
   c) Exact: $ y(0.2) = e^{-0.4} \approx 0.6703 $, error $ \approx 0.0303 $.  

6. **RK4 Method:**  
   a) $ y_{n+1} = y_n + \frac{h}{6}(k_1 + 2k_2 + 2k_3 + k_4) $.  
   b) RK4 uses multiple slope estimates per step, reducing truncation error.  

### **Section D**  
7. **Least-Squares Fitting:**  
   a) Normal equations:  
   $$
   \begin{cases}
   m \sum x_i^2 + c \sum x_i = \sum x_i y_i \\
   m \sum x_i + c n = \sum y_i
   \end{cases}
   $$  
   b) Solving gives $ m \approx 1.27 $, $ c \approx 0.86 $.  

8. **Root-Finding:**  
   a) Newton-Raphson:  
   - $ x_1 = 2 - \frac{f(2)}{f'(2)} = 2.1 $.  
   - $ x_2 = 2.1 - \frac{f(2.1)}{f'(2.1)} \approx 2.0946 $.  
   b) Bisection:  
   - Iteration 1: $ c = 2 $, $ f(c) = -1 $.  
   - Iteration 2: New interval $ [2, 3] $, $ c = 2.5 $.  
   - Iteration 3: $ c = 2.25 $.  
   c) Newton-Raphson converges quadratically; bisection converges linearly.
